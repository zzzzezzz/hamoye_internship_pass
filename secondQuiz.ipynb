```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures
sns.set()
```

```python
data = pd.read_csv('energydata_complete.csv')
data.info()
```

```python
xx = data[["T2"]]
yy = data["T6"]

Xx_train, Xx_test, yy_train, yy_test = train_test_split(xx, yy, test_size=0.3, random_state=42)
```

```python
Xx_train.info()
```

```python
lrt = LinearRegression()
lrt.fit(Xx_train, yy_train)
```

```python
lrt_pred = lrt.predict(Xx_test)
```

```python
lrt_mse = mean_squared_error(yy_test, lrt_pred)
lrt_mse
```

```python
#Q 17
#rmse for the relationship between T2 and T6

lrt_rmse = np.sqrt(lrt_mse)
lrt_rmse.round(3)
```

```python
data.isnull().sum()
```

```python
X = data.drop(columns=['Appliances', 'date', 'lights', 'rv1', 'rv2'])
y = data['Appliances']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.head()
```

```python
#normalizing my data

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#training Linear Regression model

lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

```python
#predict the model with scaled training dataset

lr_pred = lr.predict(X_train_scaled)
```

```python
#Q 18

#mae on the training set

mae = mean_absolute_error(y_train, lr_pred)
mae.round(3)
```

```python
#Q 19

# rmse on the training data

mse = mean_squared_error(y_train, lr_pred)

rmse = np.sqrt(mse)
rmse.round(3)
```

```python
#predict the model with scaled test dataset

lr_test_pred = lr.predict(X_test_scaled)
```

```python
#Q 20

#mae on the test set

test_mae = mean_absolute_error(y_test, lr_test_pred)
test_mae.round(3)
```

```python
#Q 21

# rmse on the test data

test_mse = mean_squared_error(y_test, lr_test_pred)

test_rmse = np.sqrt(test_mse)
test_rmse.round(3)
```

```python
#Q 22

#evaluating the data

r2 = r2_score(y_test, lr_test_pred)
print(f"R-sqaured: {r2:.2f}")
```

```python
#training the ridge regression with default value

ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
```

```python
#prediciting with the ridge model

ridge_pred = ridge.predict(X_test_scaled)
```

```python
#evaluating with RMSE

ridge_mse = mean_squared_error(y_test, ridge_pred)
ridge_rmse = np.sqrt(ridge_mse)
ridge_rmse.round(3)
```

```python
#Q 23

#difference in RMSE with LinearRegression and Ridge regression using test data

diff = ridge_rmse - test_rmse
diff
```

```python
#training the Lasso with default values

lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
```

```python
#Q 24

#getting the weight of the coefficients from the lasso model

weights = pd.Series(lasso.coef_, index=X_train.columns).sort_values()

# Create a DataFrame with the weights
col_name = 'Lasso_Coefficients' 
weights_df = pd.DataFrame(weights, columns=[col_name]).reset_index()
weights_df.columns = ['Features', col_name]
weights_df[col_name] = weights_df[col_name].round(3)

print(weights_df)
```

```python
#making the prediction with lasso model on the test data

lasso_pred = lasso.predict(X_test_scaled)
```

```python
#Q 25

#evaluating with RMSE

lasso_mse = mean_squared_error(y_test, lasso_pred)
lasso_rmse = np.sqrt(lasso_mse)
lasso_rmse.round(3)
```
